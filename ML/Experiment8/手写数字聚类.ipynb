```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import fowlkes_mallows_score
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" 
```

```python
digits_train = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra',
                           header=None)
digits_test = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tes',
                          header=None)
```

```python
digits_train
digits_test
```

```python
X_train = digits_train.iloc[:,:-1]
y_train = digits_train.iloc[:,-1]
X_test = digits_test.iloc[:,:-1]
y_test = digits_test.iloc[:,-1]
```

```python
X_train
y_train
X_test
y_test
```

```python
KM = KMeans(n_clusters=10)
KM.fit(X_train,y_train)
```

```python
fowlkes_mallows_score(y_test,KM.predict(X_test))
```
