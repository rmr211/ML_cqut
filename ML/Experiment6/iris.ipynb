```python
import numpy as np
import pandas as pd
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" 
```

```python
class data:
    def __init__(self,array):
        self.data_array = array
        self.data = np.array(self.data_array.iloc[:,0:4]).tolist()
        self.target = np.array(self.data_array.iloc[:,4]).tolist()
```

```python
def load_data():
    file = 'iris.csv'
    datas = pd.read_csv(file)
    return data(datas)
```

```python
datas = load_data()
```

```python
iris_x = datas.data
iris_y = datas.target
```
